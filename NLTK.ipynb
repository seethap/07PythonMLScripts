```python
def lexical_diversity(text):
    return len(set(text)) / len(text)
def percentage(count, total):
    return 100 * count / total
```

```python
# neither frequent nor infrequent words help, we need to try something else.
# fine-grain slection of words
V = set(text1)
longw =[w for w in V if len(w)>15]
sorted(longw)
```

```python
# long words are largely unique. so characterize words that are long and also occur more frequently than other long
# words.
fdist5 = FreqDist(text5)
sorted(w for w in set(text5) if len(w) > 7 and fdist5[w] > 7)
#automatically identify the frequently-occurring content-bearing words of the text.
```
